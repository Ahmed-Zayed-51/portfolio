```python
import pandas as pd
import numpy as np

# إعداد بيانات وهمية
np.random.seed(42)
num_records = 500

order_ids = range(1001, 1001 + num_records)
dates = pd.date_range(start='2025-01-01', periods=num_records)
customer_ids = [f"C{str(i).zfill(3)}" for i in range(1, num_records+1)]
product_categories = ['Electronics', 'Books', 'Clothing', 'Toys', 'Home Appliances']
products = {
    'Electronics': ['Headphones', 'Laptop', 'Smartphone', 'Camera', 'Smartwatch'],
    'Books': ['Python Guide', 'Data Science', 'Machine Learning', 'AI Basics', 'Algorithms'],
    'Clothing': ['T-Shirt', 'Jeans', 'Jacket', 'Shoes', 'Hat'],
    'Toys': ['Lego Set', 'Doll', 'Puzzle', 'Board Game', 'Toy Car'],
    'Home Appliances': ['Blender', 'Microwave', 'Vacuum', 'Air Fryer', 'Coffee Maker']
}
payment_methods = ['Credit Card', 'Debit Card', 'PayPal', 'Cash']
regions = ['North', 'South', 'East', 'West']

data = []
for i in range(num_records):
    category = np.random.choice(product_categories)
    product = np.random.choice(products[category])
    quantity = np.random.randint(1, 5)
    unit_price = np.random.randint(10, 700)
    total_price = quantity * unit_price
    payment_method = np.random.choice(payment_methods)
    region = np.random.choice(regions)
    date = dates[i]
    data.append([order_ids[i], date, customer_ids[i], category, product, quantity, unit_price, total_price, payment_method, region])

df = pd.DataFrame(data, columns=['OrderID','Date','CustomerID','ProductCategory','ProductName','Quantity','UnitPrice','TotalPrice','PaymentMethod','Region'])

# حفظ الملف على جهازك
df.to_csv("amazon_retail_sales.csv", index=False)
print("تم إنشاء CSV بنجاح في amazon_retail_sales.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'amazon_retail_sales.csv'
df = pd.read_csv(file_path)
```

```python
print(df.head())
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
df = df.dropna()
```

```python
sales_by_region = df.groupby('Region')['TotalPrice'].sum().reset_index()
print("Sales by Region:")
print(sales_by_region)
```

```python
top_products = df.groupby('ProductName')['Quantity'].sum().sort_values(ascending=False).reset_index()
print("Top Products:")
print(top_products.head(10))
```

```python
payment_method = df['PaymentMethod'].value_counts()
print("Most Used Payment Method:")
print(payment_method)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
# Daily Sales
daily_sales = df.groupby('Date')['TotalPrice'].sum().reset_index()
# Sales by Category
category_sales = df.groupby('ProductCategory')['TotalPrice'].sum().reset_index()
```

```python
# Total Sales
total_sales = df['TotalPrice'].sum()
# Average Order Value
average_order_value = df['TotalPrice'].mean()
# Unique Customers
unique_customers = df['CustomerID'].nunique()
# Purchase Frequency
purchase_frequency = len(df)/unique_customers
# Average Products per Order
average_products_per_order = df['Quantity'].mean()

print(f"Total Sales: {total_sales}")
print(f"Average Order Value: {average_order_value}")
print(f"Unique Customers: {unique_customers}")
print(f"Purchase Frequency: {purchase_frequency}")
print(f"Average Products per Order: {average_products_per_order}")
```

```python
# Daily Sales
plt.figure(figsize=(10,5))
sns.lineplot(x='Date', y='TotalPrice', data=daily_sales, marker='o')
plt.title("Daily Sales - Amazon")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()

# Top 10 Products
plt.figure(figsize=(10,5))
sns.barplot(x='Quantity', y='ProductName', data=top_products.head(10))
plt.title("Top 10 Best-Selling Products")
plt.xlabel("Quantity Sold")
plt.ylabel("Product Name")
plt.show()

# Sales by Region
plt.figure(figsize=(8,5))
sns.barplot(x='Region', y='TotalPrice', data=sales_by_region)
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```
